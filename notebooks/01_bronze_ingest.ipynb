```python
import urllib.request

# Descargar csv archivo a /Volumes/workspace/default/prueba/etl_project/bronze/landing desde internet
LANDING_PATH = "/Volumes/workspace/default/prueba/etl_project/bronze/landing"

RAW_POPULATION_PATH = f"{LANDING_PATH}/ population.csv"
POPULATION_URL = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"


RAW_CO2_PATH = f"{LANDING_PATH}/ co2.csv"
CO2_URL = "https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"



urllib.request.urlretrieve(CO2_URL, RAW_CO2_PATH)
print("✅ co2 file downloaded successfully!")


urllib.request.urlretrieve(POPULATION_URL, RAW_POPULATION_PATH)
print("✅ populations file downloaded successfully!")
```

```python
population_df = spark.read.format("csv") \
    .option("header", True) \
    .option("inferSchema", True) \
    .load(RAW_POPULATION_PATH)
display(population_df)
```

```python
co2_df = spark.read.format("csv") \
    .option("header", True) \
    .option("inferSchema", True) \
    .load(RAW_CO2_PATH)
display(co2_df)
```
